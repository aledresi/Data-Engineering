```python
import numpy as np

text1 = "mohammed anwar ali"
text2 = "ali mohammed ahmed"
bag_of_words = ("mohammed","ali","anwar","ahmed")
##text2 to vector
words_in_text1 = text1.split()
counts_list = []
for target in bag_of_words:
    current_word_count = 0
    for word in words_in_text1:
        if word == target:
            current_word_count = current_word_count + 1
    counts_list.append(current_word_count)
vector1 = np.array(counts_list)
##
##text2 to vector
words_in_text2 = text2.split()
counts_list = []
for target in bag_of_words:
    current_word_count = 0
    for word in words_in_text2:
        if word == target:
            current_word_count = current_word_count + 1
    counts_list.append(current_word_count)
vector2 = np.array(counts_list)
##

numerator = np.dot(vector1,vector2)
denominator = np.linalg.norm(vector1) * np.linalg.norm(vector2)

result = numerator / denominator

print(result)
```
