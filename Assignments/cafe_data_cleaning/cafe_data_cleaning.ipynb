### 2.	Data Type Correction

```python
data["Quantity"] = pd.to_numeric(data["Quantity"], errors="coerce")
data["Quantity"] = data["Quantity"].astype("Int64")
```

```python
data[["Price Per Unit", "Total Spent"]] = data[["Price Per Unit", "Total Spent"]].apply(pd.to_numeric, errors="coerce")
```

```python
data["Transaction Date"] = pd.to_datetime(data["Transaction Date"] , errors="coerce")
```

```python
data["Payment Method"].unique()
```

```python
data["Payment Method"] = data["Payment Method"].astype("category")
```

### 3.	Handle Missing and Invalid Values

```python
data.isna().sum()
```

```python
data["Item"].value_counts()
```

```python

data.replace(["ERROR", "UNKNOWN", "nan", "NONE"], pd.NA, inplace=True)
```

```python
data.head()
```

```python
data.tail()
```

### 4.	Data Consistency and Integrity

```python
data["Item"].value_counts()
```

```python
data["Price Per Unit"].value_counts()
```

```python
data.isna().sum()
```

```python
print(data[data['Item'] == 'Cake']['Price Per Unit'].mode()[0])
```

```python
data.loc[(data['Price Per Unit'] == 3) & (data['Item'].isna() | (data['Item'] == "")), 'Item'] = 'Cake'
```

```python
fill_items = {
    1.0: 'Cookie',
    2.0: 'Coffee',
    3.0: 'Cake',
    4.0: 'Sandwich',
    5.0: 'Salad',
}

data["Item"] = data["Item"].fillna(data["Price Per Unit"].map(fill_items))
```

```python
data['Total Spent'] = data['Total Spent'].fillna(data['Quantity'] * data['Price Per Unit'])
```

```python
data['Price Per Unit'] = data['Price Per Unit'].fillna(data['Total Spent'] / data['Quantity'])
```

```python
data['Quantity'] = data['Quantity'].fillna(data['Total Spent'] / data['Price Per Unit'])
```

```python
data['Location'] = data['Location'].fillna(data['Location'].mode()[0])
```

```python
data['Payment Method'] = data['Payment Method'].fillna(data['Payment Method'].mode()[0])
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

### 5.	New Feature adding season column

```python
season = {
    12: 'Winter', 1: 'Winter', 2: 'Winter',
    3: 'Spring', 4: 'Spring', 5: 'Spring',
    6: 'Summer', 7: 'Summer', 8: 'Summer',
    9: 'Fall', 10: 'Fall', 11: 'Fall'
}
```

```python
data['season'] = data['Transaction Date'].dt.month.map(season)
```
